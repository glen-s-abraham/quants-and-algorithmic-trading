```python
import yfinance as yf
```

```python
sectors = [
    '^NSEBANK',
    '^CNXAUTO',
    '^CNXFMCG',
    '^CNXPHARMA',
    '^CNXIT',
    '^CNXINFRA',
    '^CNXMETAL',
    '^CNXENERGY'
]
```

```python
ticker =  yf.download(tickers='^CNXFMCG', period='1y',interval='1mo')
df = ticker['Adj Close']
df = df.reset_index()
df.sort_values(by='Date')
```

```python
Date = df.iloc[:,:-1].values
Price = df.iloc[:,-1].values
```

```python
Price = Price.reshape(-1,1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(Price)
Price = sc.transform(Price)
```

```python
import matplotlib.pyplot as plt

plt.plot(Date,Price)
plt.xticks(Date.squeeze())
```
